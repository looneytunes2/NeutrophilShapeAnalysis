```python
########### DIRECTORIES ###########

#where is this analysis getting saved
savedir = 'E:/Aaron/Combined_37C_Confocal_PCA_s10/'

#which data folders to include
foldlist = ['Galvanotaxis_Confocal_40x_37C_10s/',
            'CK666_Confocal_40x_37C/',
            'ParaNitroBlebbistatin_Confocal_40x_37C/']
bigdir = 'E:/Aaron/'

########### PARAMETERS ##############
xyres = 0.3394 #um / pixel .2285
zstep = 0.7 # um
xy_buffer = 20
z_buffer = 10
time_interval = 10
smooth_factor = 10
align_method = 'trajectory'
sigma = 0
norm_rot = 'provided'
l_order = 10
nisos = [1,63]
pilr_method = 'none'
```

```python

########## GET TRAJECTORIES FROM POSITION INFO
for f in foldlist:
    if __name__ ==  '__main__':
        image_processing.get_smooth_trajectories(
            bigdir+f,
            savedir,
            time_interval,
            smooth_factor,
            )
```

```python

############ FIND WIDTH ROTATIONS THAT DEPEND ON PREVIOUS FRAMES TO LIMIT ROTATION FLIPPING ################
for f in foldlist:
    if __name__ ==  '__main__':
        image_processing.get_normal_rotations(
            bigdir+f,
            savedir,
            xyres,
            zstep,
            align_method,
            sigma,
            )

```

```python
################ MAKE MESHES FROM SEGMENTATIONS AND ALIGN THEM TO COMMON FRAME OF REFERENCE ###########
for f in foldlist:
    if __name__ ==  '__main__':
        image_processing.seg_to_mesh(
            bigdir+f, #base directory with all of the different data folders and files
            savedir,
            xyres, #xy resolution
            zstep, # z resolution
            align_method, #how to align the cells
            norm_rot, #how to perform the normal rotation
            l_order, # L order for SH coefficients
            nisos, # list of shells to calculate in PILRs
            pilr_method, #how to calculate PILRs
            sigma, #how much to smoothen image before turning to mesh
            )
```
